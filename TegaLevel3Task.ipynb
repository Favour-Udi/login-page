```python
# Importing numpy
import numpy

array_one = numpy.array([[5, 6, 9],[21, 18, 27]])  #Array one
array_two = numpy.array([[15, 33, 24],[4, 7, 1]])  #Array two

print("\nAddition of two array is\n")
addition = array_one + array_two   #Adding the two arrays and storing it in addition
print(addition)   #Printing the addition

print("\nResult array after calculating the square of all element\n")
array_squared = addition ** 2  #Squaring the sum of the arrays
print(array_squared)   #Printing the result
```

```python
#Importing matplotlib.pyplot
#Importing pandas
import pandas as pd
import matplotlib.pyplot as plt
car_sales = pd.read_csv('company_sales_data.csv', index_col =0)    #Reading the company sales data csv file
df = pd.DataFrame(car_sales)  #converting the csv file to a dataframe
print(df)   #printing the dataframe

total_montly_profit = df.loc[:,['total_profit']] #Getting the total monthly profit from the dataframe
print(total_montly_profit)  #printing the values

total_montly_profit.plot()   #plotting the line graph of the total profit per month
plt.title('Company profit per month')
plt.ylabel('Total profit')
plt.xlabel('Month Number')
plt.show()     #this enables the display of the graph
```

```python
#Importing matplotlib.pyplot, pandas, and numpy
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
car_sales = pd.read_csv('company_sales_data.csv', index_col =0) #read csv file
df = pd.DataFrame(car_sales)   #convert to dataframe

f = df['facecream'].to_numpy()  #facecream dataframe to numpy array
fc = np.sum(f)    #facecream sales sum
#print(fc)
a = df['facewash'].to_numpy()   #facewash dataframe to numpy array
fw = np.sum(a)    #facewash sales sum
t = df['toothpaste'].to_numpy()    #toothpaste dataframe to numpy array
tp = np.sum(t)     #toothpaste sales sum
b = df['bathingsoap'].to_numpy()   #bathing soap dataframe to numpy array
bs = np.sum(b)     #bathing soap sales sum
s = df['shampoo'].to_numpy()   #shampoo dataframe to numpy array
sp = np.sum(s)     #shampoo sales sum
m = df['moisturizer'].to_numpy()    #moisturizer dataframe to numpy array
mzer = np.sum(m)     #moisturizer sales sum

products = [fc, fw, tp, bs, sp, mzer]   #list of all the products
my_labels = ('FaceWash', 'FaceCream','Moisturizer', 'Shampoo', 'Bathing soap', 'Toothpaste')
plt.pie(products, labels=my_labels, autopct='%1.1f%%')     #pie plot
plt.title('Sales data')
plt.grid(True)
plt.legend()
plt.show()
```
